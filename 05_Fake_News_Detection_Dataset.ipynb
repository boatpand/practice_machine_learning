## Import Libraries

```python
# fundamental libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# for Stopword removal and Stemming
import re
import nltk
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
# for Applying countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
# for divide the dataset into train and test
from sklearn.model_selection import train_test_split
# for Passive Aggressive Classifier Algorithm
from sklearn.linear_model import PassiveAggressiveClassifier
# for measure accuracy
import sklearn.metrics as metrics
```

Import Dataset

```python
df = pd.read_csv('train.csv')
df
```

## Preprocessing

```python
# check missing values
df.info()
```

```python
# drop missing values
df = df.dropna()
```

```python
# reset its index values
messages = df.copy()
messages.reset_index(inplace = True)
messages
```

```python
# Stopword removal and Stemming
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#get the independent features
X = messages.drop('label',axis =1)
#get dependent variables
y = df['label']
```

```python
#Applying countvectorizer
#Creating the bag of words model
cv = CountVectorizer(max_features = 5000,ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
X.shape #(18285, 5000)
y =messages['label']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 0)
cv.get_feature_names()[:20]
```

```python
# get the parameters
cv.get_params()
```

```python
# The array is converted into dataframe
count_df = pd.DataFrame(X_train,columns = cv.get_feature_names())
count_df.head()
```

```python
# Passive Aggressive Classifier Algorithm
linear_clf = PassiveAggressiveClassifier(max_iter=50)
linear_clf.fit(X_train,y_train)
pred = linear_clf.predict(X_test)
score = metrics.accuracy_score(y_test,pred)
print("Accuracy: %0.3f"%score)
```
