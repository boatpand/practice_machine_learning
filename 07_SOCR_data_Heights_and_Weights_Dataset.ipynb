Data Science Project Idea: สร้างแบบจำลองการทำนายความสูงหรือน้ำหนักของมนุษย์ โดยใช้แบบจำลองการถดถอยเชิงเส้น(regression model)

# Import Libreries

```python
# Import fundamental libreries
import pandas as pd
# linear regression
from sklearn.linear_model import LinearRegression
# for split data to train & test
from sklearn.model_selection import train_test_split
# for calculate RMSE 
from sklearn.metrics import mean_squared_error
import math
```

## Import Dataset

```python
df = pd.read_excel('SOCR.xlsx')
df
```

## Preprocessing

```python
# check missing values
df.info()
```

don't have missing value!

## Preprocessing --> select features for fitting model

```python
# drop Height because this case we will predict Height
X = df.drop(['Index','Height'], axis='columns')
# set y to variable that we want to predict, in this case is Height
y = df.Height
X
```

## Model

```python
# split data train:test 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# check performane from RMSE
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```
