## Import Libraries

```python
# Import fundamental libreries
import pandas as pd
import matplotlib.pyplot as plt
# clustering with K-means
from sklearn.cluster import KMeans
# use for normalize dataset
from sklearn.preprocessing import MinMaxScaler
```

## Import dataset

```python
df = pd.read_excel('Iris.xls')
df
```

## Preprocessing

```python
# check missing values
df.info()
```

don't have missing value!

```python
# normalize dataset
scaler = MinMaxScaler()

scaler.fit(df[['Petal_width']])
df['Petal_width'] = scaler.transform(df[['Petal_width']])

scaler.fit(df[['Petal_length']])
df['Petal_length'] = scaler.transform(df[['Petal_length']])


scaler.fit(df[['Sepal_width']])
df['Sepal_width'] = scaler.transform(df[['Sepal_width']])

scaler.fit(df[['Sepal_length']])
df['Sepal_length'] = scaler.transform(df[['Sepal_length']])

df
```

we want to clustering in 3 groups by use Petal_width, Petal_length, Sepal_width, Sepal_length then check member in each clustering with Species_No & Species_name, so use k = 3

## Clustering

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Petal_width','Petal_length','Sepal_width','Sepal_length']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```
